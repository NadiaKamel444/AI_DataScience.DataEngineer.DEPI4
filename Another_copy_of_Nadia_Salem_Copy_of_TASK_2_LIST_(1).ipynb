### <span style="color:orange"><u>**Exercise 1: Creat List**<u></span>

- Ask the user to enter 5 numbers

- Store them in a list

- Print the first and last element

- Hint:
  - Use input().split() and list indexing

```python
numbers_str = input("Enter 5 numbers separated by spaces: ")
numbers = [int(num) for num in numbers_str.split()]

if len(numbers) >= 2:
    print(f"First element: {numbers[0]}")
    print(f"Last element: {numbers[-1]}")
elif len(numbers) == 1:
    print(f"Only one element entered: {numbers[0]}")
else:
    print("No numbers were entered.")
```

### <span style="color:orange"><u>**Exercise 2: Add Element**<u></span>

- Create a list of numbers
- Ask user for a number
- Add it to the list


```python
my_numbers = [10, 20, 30, 40, 50]
print(f"Original list: {my_numbers}")

user_input = input("Enter a number to add to the list: ")
try:
    new_number = int(user_input)
    my_numbers.append(new_number)
    print(f"Updated list: {my_numbers}")
except ValueError:
    print("Invalid input. Please enter a valid number.")
```

### <span style="color:orange"><u>**Exercise 3: Remove Element**<u></span>

- Create a list of fruits
- Ask user for fruit name
- Remove it if exists
- Hint:
Use append()


```python
fruits = ['Apple', 'Banana', 'Orange']
print(f"Original list: {fruits}")
f = input("Enter fruit name to remove: ")

if f in fruits:
    fruits.remove(f)
    print(f"Updated list: {fruits}")
elif f.capitalize() in fruits:
    fruits.remove(f.capitalize())
    print(f"Updated list: {fruits}")
else:
    print(f"{f} not found in the list.")
print(f"Final list: {fruits}")
```

### <span style="color:orange"><u>**Exercise 4: Check Existenc**<u></span>

- Ask user for a name

```python
names = ['Ali', 'Sara', 'Omar']
n = input("Enter name: ")

if n in names:
    print(f"{n} is in the list.")
else:
    print(f"{n} is not in the list.")
```

### <span style="color:orange"><u>**Exercise 5: Sorting**<u></span>

- Input numbers
- Sort ascending and descending


```python
numbers_str = input("Enter numbers separated by spaces: ")
try:
    numbers = [int(num) for num in numbers_str.split()]

    # Sort in ascending order
    ascending_numbers = sorted(numbers)
    print(f"Sorted in ascending order: {ascending_numbers}")

    # Sort in descending order
    descending_numbers = sorted(numbers, reverse=True)
    print(f"Sorted in descending order: {descending_numbers}")
except ValueError:
    print("Invalid input. Please enter valid numbers separated by spaces.")
```

### <span style="color:orange"><u>**Exercise 6: Remove Duplicates**<u></span>

- Convert list to unique values


```python
nums = [1, 2, 2, 3, 3, 4]
print(f"Original list: {nums}")
unique_nums = list(set(nums))
print(f"List with unique values: {unique_nums}")
```
