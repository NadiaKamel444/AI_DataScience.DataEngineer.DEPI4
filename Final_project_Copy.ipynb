```python
class Person:
    """Base class for all people in the hospital."""
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def view_info(self):
        """View basic information about the person."""
        return f"Name: {self.name}, Age: {self.age}"


class Patient(Person):
    """Class for hospital patients, inheriting from Person."""
    def __init__(self, name, age, medical_record):
        super().__init__(name, age)
        self.patient_id = id(self) # Simple unique ID for demonstration
        self.medical_record = medical_record

    def view_record(self):
        """View patient record."""
        return f"Patient ID: {self.patient_id}, Name: {self.name}, Age: {self.age}, Medical Record: {self.medical_record}"


class Staff(Person):
    """Class for hospital staff, inheriting from Person."""
    def __init__(self, name, age, position):
        super().__init__(name, age)
        self.staff_id = id(self) # Simple unique ID for demonstration
        self.position = position

    def view_info(self):
        """View staff information."""
        return f"Staff ID: {self.staff_id}, Name: {self.name}, Age: {self.age}, Position: {self.position}"


class Hospital:
    """Class for managing hospital operations."""
    def __init__(self, name, location):
        self.name = name
        self.location = location
        self.departments = []  # List to hold department objects

    def add_department(self, department):
        """Add a department to the hospital."""
        self.departments.append(department)
        print(f"Department '{department.name}' added to {self.name}.")

    def get_department(self, name):
        """Retrieve a department by name."""
        for dept in self.departments:
            if dept.name == name:
                return dept
        return None

    def display_departments(self):
        """Display all departments in the hospital."""
        if not self.departments:
            print("No departments in the hospital.")
            return
        print(f"\nDepartments in {self.name}:")
        for dept in self.departments:
            print(f"- {dept.name}")


class Department:
    """Class representing a department in the hospital."""
    def __init__(self, name):
        self.name = name
        self.patients = []  # List to hold patient objects
        self.staff = []     # List to hold staff objects

    def add_patient(self, patient):
        """Add a patient to the department."""
        self.patients.append(patient)
        print(f"Patient '{patient.name}' added to {self.name} department.")

    def add_staff(self, staff_member):
        """Add staff member to the department."""
        self.staff.append(staff_member)
        print(f"Staff '{staff_member.name}' added to {self.name} department.")

    def display_patients(self):
        """Display all patients in the department."""
        if not self.patients:
            print(f"No patients in {self.name} department.")
            return
        print(f"\nPatients in {self.name} department:")
        for patient in self.patients:
            print(f"- {patient.view_record()}")

    def display_staff(self):
        """Display all staff in the department."""
        if not self.staff:
            print(f"No staff in {self.name} department.")
            return
        print(f"\nStaff in {self.name} department:")
        for staff_member in self.staff:
            print(f"- {staff_member.view_info()}")


def hospital_ui():
    """Provides a text-based user interface for the hospital system."""
    city_hospital = Hospital("City Hospital", "123 Main St")
    print(f"Welcome to {city_hospital.name} at {city_hospital.location}!")

    while True:
        print("\n--- Hospital Management System ---")
        print("1. Add Department")
        print("2. View Departments")
        print("3. Select Department to Manage")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            dept_name = input("Enter new department name: ")
            if city_hospital.get_department(dept_name):
                print("Department already exists.")
            else:
                new_dept = Department(dept_name)
                city_hospital.add_department(new_dept)
        elif choice == '2':
            city_hospital.display_departments()
        elif choice == '3':
            dept_name = input("Enter department name to manage: ")
            selected_dept = city_hospital.get_department(dept_name)
            if selected_dept:
                manage_department_ui(selected_dept)
            else:
                print("Department not found.")
        elif choice == '4':
            print("Exiting Hospital Management System. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

def manage_department_ui(department):
    """Provides a text-based user interface for managing a specific department."""
    while True:
        print(f"\n--- Managing {department.name} Department ---")
        print("1. Add Patient")
        print("2. View Patients")
        print("3. Add Staff")
        print("4. View Staff")
        print("5. Back to Main Menu")

        choice = input("Enter your choice: ")

        if choice == '1':
            name = input("Enter patient name: ")
            age = int(input("Enter patient age: "))
            medical_record = input("Enter medical record details: ")
            patient = Patient(name, age, medical_record)
            department.add_patient(patient)
        elif choice == '2':
            department.display_patients()
        elif choice == '3':
            name = input("Enter staff name: ")
            age = int(input("Enter staff age: "))
            position = input("Enter staff position: ")
            staff = Staff(name, age, position)
            department.add_staff(staff)
        elif choice == '4':
            department.display_staff()
        elif choice == '5':
            print(f"Returning to main menu from {department.name}.")
            break
        else:
            print("Invalid choice. Please try again.")

# Run the UI
if __name__ == "__main__":
    hospital_ui()
```

```python
class BankAccount:
    def __init__(self, initial_balance=0):
        if not isinstance(initial_balance, (int, float)) or initial_balance < 0:
            raise ValueError("Initial balance must be a non-negative number.")
        self.balance = initial_balance
        print(f"Bank account created with initial balance: {self.balance}")

    def deposit(self, amount):
        if not isinstance(amount, (int, float)) or amount <= 0:
            print("Invalid deposit amount. Please enter a positive number.")
            return
        self.balance += amount
        print(f'Deposited {amount}. New balance: {self.balance}')

    def withdraw(self, amount):
        if not isinstance(amount, (int, float)) or amount <= 0:
            print("Invalid withdrawal amount. Please enter a positive number.")
            return
        if amount > self.balance:
            print('Insufficient funds. Current balance: {self.balance}')
        else:
            self.balance -= amount
            print(f'Withdrew {amount}. New balance: {self.balance}')

    def check_balance(self):
        print(f'Current balance: {self.balance}')

# Usage
print("\n--- Demonstrating BankAccount Class ---")
try:
    my_account = BankAccount(100)
    my_account.deposit(50)
    my_account.withdraw(30)
    my_account.check_balance()
    my_account.deposit(-10) # Test invalid deposit
    my_account.withdraw(200) # Test insufficient funds
    my_account.check_balance()

    # Test with invalid initial balance
    # invalid_account = BankAccount(-50)
except ValueError as e:
    print(f"Error: {e}")
```

```python
def read_txt_file(file_path):
    """Reads the contents of a text file and returns it as a string."""
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
        return content
    except FileNotFoundError:
        return f"Error: The file '{file_path}' was not found."
    except IOError:
        return "Error: An error occurred while reading the file."

class UserExtractor:
    def __init__(self, file_path):
        self.file_path = file_path
        self.usernames = {}

    def extract_usernames(self):
        """Extracts usernames from the text file and stores them in a dictionary."""
        content = read_txt_file(self.file_path)
        if content.startswith("Error:"):
            return content # Return error message if file reading failed

        lines = content.strip().split('\n')
        for line in lines:
            if ':' in line:
                username, _ = line.split(':', 1)
                self.usernames[username.strip()] = None # Store username as key, value can be anything or updated later
        return self.usernames

# Usage example
if __name__ == "__main__":
    # Create a dummy file for testing Task-6 and Task-7
    dummy_user_file_path = 'users.txt'
    with open(dummy_user_file_path, 'w') as f:
        f.write("user1:pass123\n")
        f.write("user2:password456\n")
        f.write("admin:adminpass\n")

    print("\n--- Demonstrating Task-6: read_txt_file ---")
    file_content = read_txt_file(dummy_user_file_path)
    print(f"File content:\n{file_content}")

    print("\n--- Demonstrating Task-7: UserExtractor ---")
    user_extractor = UserExtractor(dummy_user_file_path)
    extracted_users = user_extractor.extract_usernames()
    print(f"Extracted usernames: {extracted_users}")

    # Test with a non-existent file
    print("\n--- Testing UserExtractor with non-existent file ---")
    non_existent_extractor = UserExtractor('no_such_file.txt')
    print(non_existent_extractor.extract_usernames())

    # Clean up the dummy file
    import os
    os.remove(dummy_user_file_path)
```

```python
class Person:
    """Base class for all people in the hospital."""
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def view_info(self):
        """View basic information about the person."""
        return f"Name: {self.name}, Age: {self.age}"


class Patient(Person):
    """Class for hospital patients, inheriting from Person."""
    def __init__(self, name, age, medical_record):
        super().__init__(name, age)
        self.medical_record = medical_record

    def view_record(self):
        """View patient record."""
        return f"Patient Record: {self.medical_record}"


class Staff(Person):
    """Class for hospital staff, inheriting from Person."""
    def __init__(self, name, age, position):
        super().__init__(name, age)
        self.position = position

    def view_info(self):
        """View staff information."""
        return f"Staff Name: {self.name}, Age: {self.age}, Position: {self.position}"


class Hospital:
    """Class for managing hospital operations."""
    def __init__(self, name, location):
        self.name = name
        self.location = location
        self.departments = []  # List to hold departments

    def add_department(self, department):
        """Add a department to the hospital."""
        self.departments.append(department)
        print(f"Department '{department.name}' added to {self.name}.")


class Department:
    """Class representing a department in the hospital."""
    def __init__(self, name):
        self.name = name
        self.patients = []  # List to hold patients
        self.staff = []     # List to hold staff

    def add_patient(self, patient):
        """Add a patient to the department."""
        self.patients.append(patient)
        print(f"Patient '{patient.name}' added to {self.name} department.")

    def add_staff(self, staff_member):
        """Add staff member to the department."""
        self.staff.append(staff_member)
        print(f"Staff '{staff_member.name}' added to {self.name} department.")


# Example Usage
if __name__ == "__main__":
    # Create a hospital
    hospital = Hospital("City Hospital", "123 Main St")

    # Create a department
    cardiology = Department("Cardiology")
    hospital.add_department(cardiology)

    # Create a patient
    patient1 = Patient("Alice", 30, "No known allergies")
    cardiology.add_patient(patient1)

    # Create a staff member
    doctor1 = Staff("Dr. Smith", 45, "Cardiologist")
    cardiology.add_staff(doctor1)

    # View patient and staff records
    print(patient1.view_record())
    print(doctor1.view_info())
```
